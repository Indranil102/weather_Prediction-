```python
#Data preprocessing
```

```python
#drop highly corelated column
data=data.drop(columns=['maxtemp', 'temparature', 'mintemp'])
# we drop these bcz it show multi colleration
```

```python
data.head()
```

```python
print(data['rainfall'].value_counts())
```

```python
# seperate majority adn minority class
df_majority= data[data['rainfall']==1]
df_minority= data[data['rainfall']==0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
#downsample majority class to mathc minority count
df_majority_downsampled= resample(df_majority, replace=False, n_samples=len(df_minority), random_state=42)
```

```python
print(df_majority_downsampled.shape)
```

```python
df_downsampled= pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled.shape
```

```python
#shuffling the final dataframe
df_downsampled= df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_downsampled.head()
```

```python
df_downsampled['rainfall'].value_counts()
```

```python
#split features and target as X and y
x= df_downsampled.drop(columns=['rainfall'])
y= df_downsampled['rainfall']
```

```python
print(x)
```

```python
#spliiting data inot training data and test data
x_train , x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#model training
rf_model=RandomForestClassifier(random_state=42)
param_grid= {
    'n_estimators': [50,100,200],
    'max_features':["sqrt", "log2"],
    'max_depth': [10,20,30,None],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4]
}
```

```python
# Hyperparameter using GridSearchcv
grid_search= GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)
```

```python
best_rf_model= grid_search.best_estimator_
print("best parameter for Random Forest ", grid_search.best_params_)
```

```python
#model evalution
```

```python
cv_score= cross_val_score(best_rf_model, x_train, y_train, cv=5)
print("Cross validation score: ", cv_score)
print("Mean cross validation score: ", np.mean(cv_score))
```

```python
#test set performance
y_pred= best_rf_model.predict(x_test)
print("Accuracy score on test set: ", accuracy_score(y_test, y_pred))
print("Confusion matrix on test set: \n", confusion_matrix(y_test, y_pred))
print("Classification report on test set: \n", classification_report(y_test, y_pred))
```
